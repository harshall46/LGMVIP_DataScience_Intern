# Importing Required Libraries

```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from skimage import io
```

# Importing Image

```python
img = io.imread("/dog.jpg")
cv2_imshow(img)
```

 # Applying Gray Filter on Image

```python
grey_filter = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(grey_filter)
```

# Inverting the Image

```python
invert = cv2.bitwise_not(grey_filter)
cv2_imshow(invert)
```

# Applying Gaussian Blur to Image

```python
blur = cv2.GaussianBlur(invert, (21,21), 0)
cv2_imshow(blur)
```

# Inverting Blurred Image

```python
invert_blur = cv2.bitwise_not(blur)
cv2_imshow(invert_blur)
```

# Sketched Image

```python
sketch = cv2.divide(grey_filter, invert_blur, scale = 255.0)
cv2_imshow(sketch)
cv2.imwrite("Sketch.jpg", sketch)
```
